```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz

import graphviz
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = load_iris()
x_train, x_test, y_train, y_test = train_test_split(dataset.data, dataset.target)
```

```python
# Decision Tree
tree = DecisionTreeClassifier(max_depth=3)
```

```python
# Training
tree.fit(x_train, y_train)
```

```python
# Overfiting
print('Train: ', tree.score(x_train, y_train))
print('Test : ', tree.score(x_test, y_test))
```

```python
out_name = '../data/tree.dot'
export_graphviz(tree, out_file=out_name, class_names=dataset.target_names,
               feature_names=dataset.feature_names, impurity=False, filled=True)
```

```python
with open(out_name) as f:
    dot_graph = f.read()

graphviz.Source(dot_graph)
```
