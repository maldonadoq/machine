```python
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import numpy as np
```

```python
boston = load_boston()
#print(boston.DESCR)
```

```python
fig = plt.figure()

axis_x = boston.data[:,5]
axis_y = boston.target

# Bias
axis_x = np.array([np.ones(506), axis_x]).T

print(axis_x.shape)

plt.scatter(axis_x[:, 1], axis_y, alpha=0.5)
plt.xlabel('Average Number of Rooms per Dwelling')
plt.ylabel('Median Value of Owner-occupied Homes')
plt.grid()
```

Minimize Mean Square Error $\beta = \left( X^{T}X \right)^{-1}X^{T}Y$\
Rect Equation $f(x) = b + mx$

```python
axis_xT = axis_x.T
B = np.linalg.inv(axis_xT @ axis_x) @ axis_xT @ axis_y
print(B)

plt.scatter(axis_x[:, 1], axis_y, alpha=0.5)
plt.xlabel('Average Number of Rooms per Dwelling')
plt.ylabel('Median Value of Owner-occupied Homes')
plt.plot([4,9], [B[0] + B[1] * 4, B[0] + B[1] * 9], c = 'red')

plt.grid()
```
